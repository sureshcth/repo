```python
from pylab import *
a = 10.0
for a in range(5,21,5):
	th = linspace(0, pi * a/10, 200) 
	x = a * cos(th)
	y = a * sin(th)
	plot(x,y)
```

```python
from pylab import *
N = 100 # number of points
x = linspace(0.0, 2 * pi, N) 
y = zeros(N)
for n in range(5):
	term = sin((2*n+1)*x) / (2*n+1)
	y = y + term
	plot(x,y)
```
